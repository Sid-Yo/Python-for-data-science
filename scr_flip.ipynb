```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_soup(url):
    try:
        page=requests.get(url)
        if page.status_code==200:
            return BeautifulSoup(page.text,'lxml')
        else:
            print('error',page.status_code)
    except Exception as e:
        print(e)
```

```python
soup=get_soup("https://www.flipkart.com/search?q=bags&sid=reh%2C4d7%2Cak9&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_4_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_4_na_na_na&as-pos=2&as-type=RECENT&suggestionId=bags%7CBackpacks&requestId=5a6a3309-9498-46f4-b765-df2bc21e40f4&as-searchtext=bags")
```

```python
soup.title.text
```

```python
area=soup.find_all('div',attrs={'class':"_1HmYoV _35HD7C"})[1]
```

```python
rows=area.find_all("div",attrs={'class':'_3O0U0u'})
```

```python
result=[]
for row in rows:
    bags = row.find_all('div',attrs={'class':'IIdQZO _1SSAGr'})
    for bag in bags:
        brand=bag.find('div',attrs={'class':"_2B_pmu"}).text
        title=bag.find('a',attrs={'class':"_2mylT6"}).text
        colour=bag.find('div',attrs={'class':"zU9Of8"}).text
        price=bag.find('div',attrs={'class':"_1vC4OE"}).text
        dis=bag.find('div',attrs={'class':"VGWI6T"}).span.text
        result.append({'brand':brand,"title":title,"colour":colour,"price":price[1:].replace(',',''),"discount":dis.split()[0][:-1]})
```
